```python
import pandas as pd
```

```python
df = pd.read_csv('poker_player_rank_raw.csv')
```

```python
#take a look at our dataframe (we have no CHANGE value)
df
```

```python
df['RANK'] = df['RANK'].str[1:]
```

```python
df['RANK'] = pd.to_numeric(df['RANK'])
```

```python
df = df[df['PLAYER'] != 'Unknown Player']
```

```python
start_year = 2019

while start_year != 2012:

    current_df = df[df['YEAR'] == start_year]
    prev_df = df[df['YEAR'] == start_year-1]

    for player in current_df['PLAYER']:
        try: 
            
            current_rank = current_df.loc[current_df['PLAYER'] == player, 'RANK'].iloc[0]
            former_rank = prev_df.loc[prev_df['PLAYER'] == player, 'RANK'].iloc[0]

            diff = former_rank - current_rank
            
            if diff > 0:
                diff = f"+{diff}"
            

            df["CHANGE"][(df["PLAYER"] == player) & (df["YEAR"] == start_year)] = diff

        except:
            df["CHANGE"][(df["PLAYER"] == player) & (df["YEAR"] == start_year)] = 'NEW'
            
    start_year = start_year-1
```

```python
df['PLAYER'] = df['PLAYER'].str.strip()
df['PLAYER'].str.len()
```

```python
df[df['PLAYER'] == 'Hungrybox']
```

```python
df['RANK'] = df['RANK'].astype('int64')
```

```python
df[(df["YEAR"] == 2015) & (df['RANK'] <= 20)]
```

```python
#drop all rows that contain a value of 2013 for YEAR
df.drop(df[df['YEAR'] == 2013].index, inplace=True)
```

```python
df['IS_NEW'] = df['CHANGE'] == 'NEW'
```

## a dataframe for all the NEW players on the rankings:

```python
df_of_new_players = df[df['CHANGE'] == 'NEW']
```

```python
df_of_new_players
```

```python
df['CHANGE'] = df['CHANGE'].replace('NEW', 101 - df['RANK'])
```

```python
df_of_new_players['AVG_NEW_RANK'] = df_of_new_players.groupby('YEAR')['RANK'].transform('mean')
```

```python
num_of_players = df_of_new_players.groupby('YEAR')['PLAYER'].count()
mean_of_news = df_of_new_players.groupby('YEAR')['AVG_NEW_RANK'].mean()
```

```python
a = pd.Series(num_of_players, name="NEW_ARRIVALS")
dfa = a.to_frame()

b = pd.Series(mean_of_news, name="AVG_RANK")
dfb = b.to_frame()

meta_df = pd.merge(dfa, dfb, on="YEAR")
```

```python
meta_df.reset_index(inplace=True)
```

#### We've got meta_df which contains the number of new arrivals per year and the average rank of those arrivals.
